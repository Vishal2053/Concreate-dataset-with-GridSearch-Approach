```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/concrete.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
 # Create a MinMaxScaler instance
scaler = MinMaxScaler()
 # Fit and transform the selected columns
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
 # Save the normalized data back to a CSV file
df.to_csv('normalized_data.csv', index=False)
```

```python
data=pd.read_csv('normalized_data.csv')
data.head()
```

```python
X=data[['cement', 'slag', 'ash', 'water', 'superplastic', 'coarseagg',
       'fineagg', 'age']]
y=data[['strength']]
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object.
encoder = LabelEncoder()

# Fit the LabelEncoder object to the labels.
encoder.fit(y)

# Convert the continuous labels to categorical labels.
encoded_labels = encoder.transform(y)

# Print the encoded labels.
print(encoded_labels)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,encoded_labels,test_size=0.4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
model_params={
  'svm':{
  'model':svm.SVC(gamma='auto'),
  'params':{
  'C':[1,10,20],
  'kernel':['rbf','linear']
  }
  },
  'random_forest':{
  'model':RandomForestClassifier(),
  'params':{
  'n_estimators':[1,5,10]
  }
  },
  'logistic_regression':{
  'model':LogisticRegression(solver='liblinear',multi_class='auto'),
  'params':{
  'C':[1,5,10]
  }
  }
}
```

```python
scores=[]
for model_name,mp in model_params.items():
  GS=GridSearchCV(mp['model'],mp['params'],cv=2,return_train_score=False)
  GS.fit(X_train,y_train)
  scores.append({
    'model':model_name,
    'best_score':GS.best_score_,
    'best_params':GS.best_params_
  })
pd.DataFrame(scores,columns=['model','best_score','best_params'])
```

```python
from sklearn.svm import SVC
model=SVC(C=1,kernel='rbf')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
model.score(X_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier(n_estimators=100)
model1.fit(X_train,y_train)
```

```python
ypred=model1.predict(X_test)
```

```python
model1.score(X_test,ypred)
```
